```python
import pandas as pd
Data = pd.read_csv("LinkedIn Profile Data.csv", low_memory=False)
Data.info()

Data['if_male'] = (Data['gender'] == 'Male') * 1
Data.dropna(inplace=True)
print("missing values before filling:")
Data.isna().any()
```

```python
print(Data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

avg_time_in_previous_position = Data['avg_time_in_previous_position']

plt.hist(avg_time_in_previous_position, bins=20, range=(0, 8), edgecolor='black')  # Limiting range from 0 to 8 years
plt.xlabel('Average Time in Previous Positions in years')
plt.ylabel('Frequency')
plt.title('Histogram of Average Time Spent in Previous Positions')
plt.grid(True)
plt.show()
```

```python
age_gender_data = Data[['age', 'if_male']]  # Considering 'if_male' as a column for gender

female_ages = age_gender_data[age_gender_data['if_male'] == 0]['age']
male_ages = age_gender_data[age_gender_data['if_male'] == 1]['age']

# Plotting boxplots for age distribution based on gender
plt.figure(figsize=(8, 6))
plt.boxplot([female_ages.dropna(), male_ages.dropna()], labels=['Female', 'Male'])
plt.title('Distribution of Ages by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
data_with_promotion = Data[['no_of_promotions', 'gender']]

plt.figure(figsize=(8, 6))

# Scatter plot for female beauty vs number of promotions
plt.scatter(data_with_promotion['gender'], data_with_promotion['no_of_promotions'])

plt.ylabel('Number of Promotions')
plt.xlabel('gender')
plt.title('Number of Promotions vs Gender')
```
